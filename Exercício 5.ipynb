## Exercício 5

**Importando bibliotecas**

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

**Importando dados como um dataframe**

```python
dataframe = pd.read_fwf('data.txt',header=None,
names=["Router1", "Router2", "Router3", "Router4","Router5","Router6","Router7", "Room"])
```

**Criando função para plotar gaussiana**

```python
def plotGaussian(df):
    
    mean = df.mean() # Média do dataset
    std = df.std() # Desvio padrão do dataset
    
    # Gerando uma distruibuição normal com os dados do dataset
    x_values = np.linspace(df.min(), df.max(), 400)
    y_values = stats.norm.pdf(x_values, mean, std)
    
    plt.plot(x_values, y_values)
```

---

### Roteador 1

```python
ex5Router1 = dataframe.loc[:, 'Router1'] # Isolando o roteador 
ex5Router1.hist(bins=15, density=1) # Gerando um histograma com frequência relativa
ex5Router1.plot.density()
plotGaussian(ex5Router1) # Verde - Gerando uma gaussiana 
stats.shapiro(ex5Router1)
```

#### Resposta
Podemos observar pelo gráfico que os dados do roteador 1 não seguem uma distribuição gaussiana <br>
Para comprovar o resultado foi utilizado o shapiro<br>
Portanto para os níveis de significância de: <br>
1% -> Rejeitamos o H0 <br>
5% -> Rejeitamos o H0 <br>
10% -> Rejeitamos o H0 <br>
**Portanto podemos concluir que o roteador 1 não segue uma distribuição gaussiana**
